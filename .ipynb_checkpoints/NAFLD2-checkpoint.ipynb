```python
data = pd.read_excel("NAFLD.xlsx")
features = data.iloc[:,1:-7]
targets = data.iloc[:,-7:]
```

```python
new_fibrosis = []
for fibrosis in np.array(targets.iloc[:,1:5]):
    if fibrosis[3] == 1:
        new_fibrosis.append(4)
    elif fibrosis[2] == 1:
        new_fibrosis.append(3)
    elif fibrosis[1] == 1:
        new_fibrosis.append(2)
    else:
        new_fibrosis.append(fibrosis[0])
```

```python
new_ = np.array(new_fibrosis)[:, np.newaxis]
targets.iloc[:,1:5] = new_
```

```python
new_targets = targets.drop(labels=list(targets.columns[1:4]), axis=1)
# new_targets.iloc[:,1] = new_targets.iloc[:,1].map({0:2, 1:3, 2:4, 3:5, 4:6})
# new_targets.iloc[:,2] = new_targets.iloc[:,2].map({1:7, 2:8})
# new_targets.iloc[:,3] = new_targets.iloc[:,3].map({1:9, 2:10})
```
